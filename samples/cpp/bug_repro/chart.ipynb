```python
import csv
import matplotlib.pyplot as plt
```

```python
def get_rss_data(filename, start_with=0, max_iter=-1):
    x_data = []
    with open(filename, "r") as csv_file:
        csv_reader = csv.reader(csv_file)
        count = 0
        for row in csv_reader:
            count += 1
            if count < start_with:
                continue
            elif max_iter != -1 and count >= max_iter:
                break
            x_data.append(float(row[0]) / 1024 ** 2)
    return x_data

def get_unique_words_count_data(filename, start_with=0, max_iter=-1):
    unique_words = set()
    count = 0
    x_data = []
    with open(filename, 'r') as file:
        for line in file:
            words = line.strip().split()
            unique_words.update(words)
            count += 1
            if count < start_with:
                continue
            elif max_iter != -1 and count >= max_iter:
                break
            x_data.append(len(unique_words))
    return x_data

def print_chart(filename, start_with=0, max_iter=-1):
    x_data = []
    with open(filename, "r") as csv_file:
        csv_reader = csv.reader(csv_file)
        count = 0
        for row in csv_reader:
            count += 1
            if count < start_with:
                continue
            elif max_iter != -1 and count >= max_iter:
                break
            x_data.append(float(row[0]))
    
    plt.plot(x_data)
    plt.xlabel("InferIdx")
    plt.ylabel("RSS")
    plt.title(filename)
    plt.show()

def print_unique_words_count(filename, start_with=0, max_iter=-1):
    unique_words = set()
    count = 0
    x_data = []
    with open(filename, 'r') as file:
        for line in file:
            words = line.strip().split()
            unique_words.update(words)
            count += 1
            if count < start_with:
                continue
            elif max_iter != -1 and count >= max_iter:
                break
            x_data.append(len(unique_words))
    plt.plot(x_data)
    plt.xlabel("InferIdx")
    plt.ylabel("Unique shapes count")
    plt.title(filename)
    plt.show()
```

```python
rss_data = get_rss_data("test_300.csv")
unique_words_count_data = get_unique_words_count_data("shapes_300.txt")

fix, (ax1, ax2) = plt.subplots(2, 1, figsize=(6,8))
ax1.plot(rss_data)
ax1.set_title("Memory consumption")
ax1.set_xlabel("Infer Index")
ax1.set_ylabel("RSS, GB")

ax2.plot(unique_words_count_data)
ax2.set_title("Unique input shapes count")
ax2.set_xlabel("Infer Index")
ax2.set_ylabel("Count")
plt.tight_layout()
plt.show()
```
